## K-Means algorithm to divide the cars in 5 customer segments

### making blobs for K-Means

```python
features, true_labels = make_blobs(
    n_samples=205,
    centers=5,
    cluster_std=2.75,
    random_state=42
)
```

```python
features
```

```python
kmeans = cluster.KMeans(
    init="random",
    n_clusters=5,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

### making the list for all 205 cars, taking into account the average of city and highway mileage, horsepower, price and carbody type

```python
features = [[0, 0, 0, 0, 0, 0, 0, 0] for i in range(205)]
# print(len(features))
for i in range(205):
    features[i][0] = (df.iat[i, 22] + df.iat[i, 23]) / 2 # mileage
    features[i][1] = df.iat[i, 20] # horsepower
    features[i][2] = df.iat[i, 24] # price
    features[i][3] = df.iat[i, 31] # convertible
    features[i][4] = df.iat[i, 32] # hardtop
    features[i][5] = df.iat[i, 33] # hatchback
    features[i][6] = df.iat[i, 34] # sedan
    features[i][7] = df.iat[i, 35] # wagon
features
```

### scaling the features before running k-means

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features
```

### running k-means

```python
kmeans.fit(scaled_features)
```

### these centers are the random 5 random centroids chosen by k-means algorithm

```python
kmeans.cluster_centers_
```

### kmeans labels represent the labels assigned to each car, each label corresponds to one customer segment

```python
kmeans.labels_
```

```python
df.at[:, 'cluster'] = kmeans.labels_
```
