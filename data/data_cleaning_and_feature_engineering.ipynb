## Feature Engineering

```python
mil_low = [0, 20, 30]
mil_high = [30, 40, 60]
mil_med = [20, 30, 40]

hp_low = [40, 80, 140]
hp_high = [140, 200, 300]
hp_med = [80, 140, 200]

price_low = [4000, 12000, 25000]
price_high = [25000, 35000, 50000]
price_med = [15000, 25000, 35000]

def get_low(low, val):
    # low is a 3 tuple, val is the value to be classified
    if val > low[2]:
        return 0
    elif val < low[1]:
        return 1
    else:
        return (low[2] - val)/(low[2] - low[1])
def get_high(high, val):
    if(val < high[0]):
        return 0
    elif val > high[1]:
        return 1
    else:
        return (val - high[0])/(high[1] - high[0])
    
def get_med(med, val):
    if(val < med[0] or val > med[2]):
        return 0
    if(val < med[1]):
        return (val - med[0])/(med[1] - med[0])
    else:
        return (med[2] - val)/(med[2] - med[1])
    
def membership(row):
    # row is a 3 tuple {mileage, hp, price}
    
    M = [0] * 9 # {mil 3 tuple}, {hp 3 tuple}, {price 3 tuple}
    
    for i in range(0, 3):
        M[0] = get_low(mil_low, row[0])
        M[1] = get_med(mil_med, row[0])
        M[2] = get_high(mil_high, row[0])
    
    for i in range(0, 3):
        M[3] = get_low(hp_low, row[1])
        M[4] = get_med(hp_med, row[1])
        M[5] = get_high(hp_high, row[1])
    
    for i in range(0, 3):
        M[6] = get_low(price_low, row[2])
        M[7] = get_med(price_med, row[2])
        M[8] = get_high(price_high, row[2])
    
    return M

print(*(membership([27, 100, 16000])))
    
    
    
```

```python
cnt = 0
for i in range(0, 65):
    Membership = membership([(df.iat[i, 22] + df.iat[i, 23])/2, df.iat[i, 20], df.iat[i, 24]])
    
    if df.iat[i, 5] == 'hatchback' or df.iat[i, 5] == 'sedan':
        # high mileage > 0.3, low price > 0.4
        if(Membership[2] > 0.3 and Membership[6] > 0.4):
            df.at[i, 'Customer Segment'] = 'Middle Class'
            
    if df.iat[i, 5] == 'sedan' or df.iat[i, 5] == 'hardtop':
        # med mileage > 0.2 or high mileage > 0.1, med hp > 0.2
        if((Membership[1] > 0.2 or Membership[2] > 0.1) and Membership[4] > 0.2):
            df.at[i, 'Customer Segment'] = 'Upper Middle Class'
    
    if df.iat[i, 5] == 'convertible' or df.iat[i, 5] == 'sedan':
        # high hp > 0.2
        if(Membership[5] > 0 or Membership[4] > 0.3):
            df.at[i, 'Customer Segment'] = 'Highclass Youngsters'
            
    if df.iat[i, 5] == 'wagon' or df.iat[i, 5] == 'sedan':
        # med mileage > 0.2, med hp > 0.5 or high hp > 0.1
        if(Membership[1] > 0.2 and (Membership[4] > 0.5 or Membership[5] > 0.1)):
            df.at[i, 'Customer Segment'] = 'Highclass Families'

            
df.to_excel('data_set_after_feature_engineering.xlsx')
```
